```python
# model name: rf_model
# dataset name: X_m (pandas)

importances = rf_model.feature_importances_
colnames = X_m.columns
values = importances
```

```python
#This creates a table containing values for variable importance
df_m = pd.DataFrame({"variable": colnames, "importance": list(values)})
df_m_sorted = df_m.sort_values(by="importance", ascending=False)
display(df_m_sorted) #this display the table
```

```python
sns.barplot(data=df_m_sorted, y="variable", x="importance", orient = "h",
            palette = sns.light_palette("seagreen", 15, reverse = True)[2:])
ax = plt.gca()
for i, value in enumerate(df_m_sorted["importance"]):
    ax.text(value+0.002, i, f"{value:.3f}", va='center', color='#22b600', 
            fontweight = "bold", fontname='Times New Roman', fontsize = 12)
plt.xlim([0.0, 0.305])
plt.suptitle('Variable importance, sample male population', fontsize = 14, fontname='Times New Roman',
             fontweight='heavy')
```
